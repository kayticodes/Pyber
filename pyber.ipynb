```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load 
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"
```

```python
# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)
```

```python
# Display city data table for preview
city_df.head()
```

```python
# Display Ride data table for preview
ride_df.head()
```

```python
# Combine the data into a single dataset
merged_df = pd.merge(ride_df, city_df, on="city")
# Display the data table for preview
merged_df
```

```python
#See how many unique citys there are and how many times they come up 
city_count= merged_df.groupby[['city']].nunique()
city_count
```
